Нарисуйте график функци: y(x) = k*cos(x-a)+b для некоторых (2-3 различных) значений k,a,b

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
x = np.linspace(-np.pi, np.pi, 201)
def y(x,k,a,b):
    return k*np.cos(x-a)+b
k,a,b = 1,3,4
plt.plot(x,y(x,k,a,b))
k,a,b = -1,2,2
plt.plot(x,y(x,k,a,b))
k,a,b = 3,2,-1
plt.plot(x,y(x,k,a,b))
```
