Напишите код, который будет переводить полярные координаты в дакертовы

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
def p2c (R,a):
  x = R* np.cos(np.radians(a))
  y = R*np.sin(np.radians(a))
  return x,y
print (p2c (15,45))
```

напишите код, который будет рисовать график окружности в полярных координатах:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
def pcircle  (R):
    phi = np.linspace(0,2*np.pi,201)
    r = np.linspace(R,R,201)
    plt.polar(phi,r)
    plt.show()
pcircle(1)
```

  Напишите код, который будет рисовать график отрезка прямой линии в полятных координатах:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
def pline  (L, R):
    phi = np.array([np.radians(L), np.radians(L)])
    r = np.array([0,R])
    plt.polar(phi,r)
    plt.show()
pline(75,5)
```
