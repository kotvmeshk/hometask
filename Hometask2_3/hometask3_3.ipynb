```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
def my_circle(r, x0=0, y0=0):
    x = []
    y = []
    for t in np.linspace(0, 360, 361):
        x.append(r * np.cos((np.pi * t) / 180) + x0)
        y.append(r * np.sin((np.pi * t) / 180) + y0)
    plt.plot(x, y)
    plt.plot(x0, y0, marker='o')
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.axis('equal')
    plt.show()

my_circle(4, -1, 2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
def my_elipse(a, b, x0=0, y0=0):
    x = []
    y = []
    for t in np.linspace(0, 360, 361):
        x.append(a * math.cos((np.pi * t) / 180) + x0)
        y.append(b * math.sin((np.pi * t) / 180) + y0)
    plt.plot(x, y)
    plt.plot(x0, y0, marker='o')
    plt.axis('equal')
    plt.show()

my_elipse(2, 1, -1, 1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
def my_gip(a, b, x0=0, y0=0):
    x1 = []
    x2 = []
    y = []
    for t in np.linspace(-3, 2, 100):
        x1.append(a * math.cosh(t) + x0)
        x2.append(-a * math.cosh(t) + x0)
        y.append(b * math.sinh(t) + y0)
    plt.plot(x1, y)
    plt.plot(x2, y)
    plt.plot(x0, y0, marker='o')
    plt.show()

my_gip(3, 4, -2, 3)
```
