```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
t = np.linspace(0,2 *np.pi,100)
plt.figure(figsize = (4,4))
plt.plot(np.cos(t),np.sin(t))
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
u =1
v=0.5
a =2
b=1.5
t = np.linspace(0,2*np.pi, 100)
plt.plot(u+a*np.cos(t), v+b*np.sin(t))
plt.show()
```
