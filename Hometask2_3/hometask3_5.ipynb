```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig =figure()
ax = Axes3D (fig)
X =np.arange(-3,5,0.5)
Y =np.arange(-1,3,0.5)
X,Y =np.meshgrid(X,Y)
Z = np.sqrt(X**2 -Y**2)

ax.plot_wireframe (X,Y,Z)
show()
```
